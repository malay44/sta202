### Fitting Normal Distribution to get Mean and SD

```python
from scipy.stats import norm
import numpy as np

# Assuming 'height_data' contains the height data of children
height_data = np.array(child_height_data)

# Fit a normal distribution using maximum likelihood estimation
mu, std = norm.fit(height_data)

# Print parameters of the fitted normal distribution
print("Mean:", mu)
print("Standard Deviation:", std)
```

### Hypothesis Testing(One Sample Mean Test): To Test if the Average height of Child differs significantly from the mean.

```python
from scipy.stats import ttest_1samp
height_data = np.array(child_height_data)
# Assuming we want to test if the average height differs significantly from a certain value (e.g., 170 cm)
null_hypothesis_mean = 163.91148148148147

# Perform one-sample t-test
t_statistic, p_value = ttest_1samp(height_data, null_hypothesis_mean)

# Print test results
print("One-sample t-test results:")
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Decision based on p-value
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: The average height differs significantly from", null_hypothesis_mean)
else:
    print("Fail to reject null hypothesis: The average height does not differ significantly from", null_hypothesis_mean)
```

### Hypothesis Testing(Two Sample Mean Test): To Test if the Average height of Child differs significantly from height of father and mother.

```python
import pandas as pd
df = pd.read_csv('height_data.csv')
# Assuming we want to test if there's a significant difference in heights between children of different genders
father_heights = df['father_height']
mother_heights = df['mother_height']
child_heights = df['height']
```

#### Father and Child

```python
from scipy.stats import ttest_ind

# Perform two-sample t-test
t_statistic, p_value = ttest_ind(child_heights, father_heights)

# Print test results
print("Two-sample t-test results:")
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Decision based on p-value
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There's a significant difference in heights between child and father height.")
else:
    print("Fail to reject null hypothesis: There's no significant difference in heights between child and father height.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot between height of child and height of father
plt.figure(figsize=(8, 6))
sns.scatterplot(x=child_heights, y=father_heights)
plt.title('Scatter Plot between Height of Child and Height of Father')
plt.xlabel('Height of First Child')
plt.ylabel('Height of Second Child')
plt.show()

# Compute correlation coefficient
correlation_coefficient = child_heights.corr(father_heights)
print("Correlation Coefficient:", correlation_coefficient)
```

```python
import statsmodels.api as sm

# Define independent variable (height of child)
X = child_heights

# Add a constant term for the intercept
X = sm.add_constant(X)

# Define dependent variable (height of father)
y = father_heights

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```

```python
# Get predicted values and confidence intervals
pred_results = model.get_prediction(X)

# Extract predicted values
pred_values = pred_results.predicted_mean

# Extract confidence intervals
conf_int = pred_results.conf_int()

import matplotlib.pyplot as plt

# Scatter plot between height of child and height of mother
plt.figure(figsize=(8, 6))
plt.scatter(child_heights, father_heights, label='Data Points')

# Plot regression line
plt.plot(child_heights, pred_values, color='red', label='Regression Line')

# Plot confidence intervals
plt.fill_between(child_heights, conf_int[:, 0], conf_int[:, 1], color='blue', alpha=0.3, label='95% Confidence Interval')

plt.title('Linear Regression with Confidence Intervals')
plt.xlabel('Height of Child')
plt.ylabel('Height of Father')
plt.legend()
plt.show()
```
